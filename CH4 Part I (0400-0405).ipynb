# show() or no show()?

## plotting from a script
if using matplotlib from within a script the function plt.show is best, it starts an event loop looks for all active figure objects and open one or more interactive windows that display the figures

one thing to be aware of: the plt.show() command should be used only once per python session, it's most often seen at the end of script. multiple show() commands can lead to unpredictable backend-dependent behaior and should be mostly avoided.


## plotting from an ipython notebook
Mostly what I do-- Plotting interactively within an ipython notebook can be done within the %matplotlib command, and works in a similar way to the ipython shell. two options are:

%matplotlib notebook will lead to interactive plots embedded within the notebook
%matplotlib inline will lead to static image of your plot embedded in the notebook
mostly use second option

### object-oriented interface is better 

```python
#first creat a grid of plots
#ax will be an array of two axes objects
fig, ax= plt.subplots(2)

#call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))


#The MATLAB-style tool are contained in the pyplot interface
#create a plot figure
plt.figure()
# create the first of two panels and set current axis
plt.subplot(2,1,1) #rows, columns, panel number
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));

#plt.plot() is MATLAB style
#ax.plot() is object oriented style
```

# 0401 Simple Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
#start by creating a figure with axes
fig=plt.figure()
ax = plt.axes()
```

In Matplotlib, the figure can be thought of as a single container that contains all the objects representing axes, graphics, text, and labels. The axes is what we see above, a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization. Throughout this book, we will commonly use the variable name fig to refer to a fiture instance, and ax to refer to an axes instance or group of axes instances.
Once we created an axes, we can use the ax.plot function to plot some data

```python
fig=plt.figure()
ax = plt.axes()

x= np.linspace(0,10,1000) #creates the x axis
ax.plot(x, np.sin(x))
```

```python
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

## Adjusting the plot: axes limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5,1.5)
```

```python
#reverse the axes
plt.plot(x, np.sin(x))

plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

## Labeling plots
Titile, axis labels and simple legends

```python
plt.plot(x, np.sin(x))
plt.title ('A sine curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

MATPLOTLIB plt vs ax

    plt.xlabel() → ax.set_xlabel()
    plt.ylabel() → ax.set_ylabel()
    plt.xlim() → ax.set_xlim()
    plt.ylim() → ax.set_ylim()
    plt.title() → ax.set_title()
    
    
And it's more convenient to use ax.set() method to set all these properties at once:

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```

# 0402 Simple Scatter Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

```python
plt.plot(x, y, '-h', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

A second, more powerful methodl is plt.scatter function

```python
plt.scatter(x,y, marker='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```
